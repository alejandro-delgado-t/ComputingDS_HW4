```python
import sys
import os
cwd = os.getcwd()
src_path = os.path.abspath(os.path.join(cwd, '..', 'src'))
sys.path.append(src_path)
```

```python
from diabetes_prediction_package.model_function import load_data, split_train_test, remove_na, na_impute_mean, encode_ethnicity, binary_gender, train_model, predict, accuracy_sc, predict_probababilities, calc_roc_auc
```

```python
## part a
df = load_data()


## part c
df = remove_na(df)


## part d
df = na_impute_mean(df)


## part e
df = encode_ethnicity(df)


## part f
df = binary_gender(df)
```
